```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('StudentsPerformance.csv')
```

```python
df
```

```python
# Visualization 1: Histogram of Math Scores
plt.figure(figsize=(7, 5))
sns.histplot(df["math score"], bins=10, kde=True, color="Lightblue")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.title("Math Scores")
plt.grid(True)
plt.show()
```

```python
# Visualization 2: Box Plot of Math Scores by Gender
plt.figure(figsize=(9, 6))
sns.boxplot(x='gender', y='math score', data=df, palette='coolwarm')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.title('Math Scores by Gender')
plt.grid(True)
plt.show()
```

```python
# Visualization 3: Bar Chart of Parental Education Levels
plt.figure(figsize=(10,6))
sns.countplot(x='parental level of education', data=df, hue="parental level of education" ,palette="viridis")
plt.xlabel('Education Level')
plt.ylabel('N.o of parents')
plt.title('Bar Chart of Parental Education Levels')
plt.grid(True)
plt.show()
```

```python
# Visualization 4: Scatter Plot of Reading vs. Writing Scores
plt.figure(figsize=(7, 5))
sns.scatterplot(x="writing score", y="reading score", data=df, hue="gender", palette="plasma", alpha=0.7)
plt.xlabel("writing Score")
plt.ylabel("reading Score")
plt.title("Writing vs.Reading Scores")
plt.legend(title="Gender")
plt.show()
```

```python
# Visualization 5: Pie Chart of Genders
plt.figure(figsize=(7, 5))
df['gender'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightpink', 'grey'])
plt.title("Genders")
plt.show()
```
